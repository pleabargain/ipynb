```python
from graphviz import Graph, Digraph

g = Digraph('G', engine="circo")  

mydict = {"the hotel is": ["old",
                  "terrible", 
                   "terrific", 
                   "cheap",
                   "affordalble",
                   "remote",
                   "in the center of the city",
                   "on the coast",
                   "in the mountains",
                   "difficult to find",
                   "new", 
                   "amazing", 
                   "tired",
                   "rundown", 
                   "huge",
                   "small",
                   "quaint",
                   "cute",
                   "unloved", 
                   "unfriendly", 
                   
                  ]}

key = "the hotel is"
values = mydict["the hotel is"]

for v in values:
    g.edge(key, v)
g
```

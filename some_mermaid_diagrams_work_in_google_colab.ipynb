# works with flow charts

````python
import base64
import requests, io
from PIL import Image
import matplotlib.pyplot as plt

graph = """
graph LR;
    A--> B & C & D;
    B--> A & E;
    C--> A & E;
    D--> A & E;
    E--> B & C & D;
"""

graphbytes = graph.encode("ascii")
base64_bytes = base64.b64encode(graphbytes)
base64_string = base64_bytes.decode("ascii")
img = Image.open(io.BytesIO(requests.get('https://mermaid.ink/img/' + base64_string).content))
plt.imshow(img)
````

# What's for dinner flow chart

````python
import base64
import requests, io
from PIL import Image
import matplotlib.pyplot as plt

graph = """
graph TD
   A(what's for dinner) --> B{fridge has food?}
   B -->|No| H(go shopping)
   B -->|Yes| C{leftovers?} -->|Yes| G(heat up)
   C -->|No| D{eggs?} -->|Yes| I(boil eggs)
   D -->|No| H(go shopping)
"""

graphbytes = graph.encode("ascii")
base64_bytes = base64.b64encode(graphbytes)
base64_string = base64_bytes.decode("ascii")
img = Image.open(io.BytesIO(requests.get('https://mermaid.ink/img/' + base64_string).content))
plt.imshow(img)
````
